```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img,'ABCDE',org=(50,300),fontFace = font,fontScale = 5,color = (255,255,255),thickness = 25,lineType = cv2.LINE_AA)
    return blank_img
```

```python
def show_img(img) :
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = 'gray')
```

```python
img = load_img()
show_img(img)
```

```python
#erosion
```

```python
kernel = np.ones((5,5),dtype = np.uint8)
erosion1 = cv2.erode(img,kernel,iterations=1)
show_img(erosion1)
```

```python
kernel = np.ones((5,5),dtype = np.uint8)
erosion1 = cv2.erode(img,kernel,iterations=4)
show_img(erosion1)
```

```python
#Opening
```

```python
img = load_img()
white_noise = np.random.randint(low=0,high=2,size = (600,600))
white_noise
```

```python
white_noise = white_noise * 255
white_noise.shape
white_noise
```

```python
noise_img = white_noise + img
show_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
show_img(opening)
```

```python
show_img(img)
```

```python
#closing
img = load_img()
black_noise = np.random.randint(low=0,high=2,size = (600,600))
black_noise
```

```python
black_noise = black_noise * -255
black_noise = black_noise + img
black_noise
```

```python
show_img(black_noise)
```

```python
black_noise[black_noise == -255] = 0
show_img(black_noise)
```

```python
closing = cv2.morphologyEx(black_noise,cv2.MORPH_CLOSE,kernel)
show_img(closing)
```

```python
img = load_img()
show_img(img)
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
show_img(gradient)
```
