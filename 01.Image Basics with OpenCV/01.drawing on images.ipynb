```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
blank_img = np.zeros(shape = (512,512,3),dtype = np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,(384,0),(510,128),color = (0,255,0),thickness = 5)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,(200,0),(0,200),(255,0,0),5)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,(100,500),(0,400),(0,0,255),5)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,(100,100),50,(255,255,255),5)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,(300,300),50,(255,0,0),-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,(0,0),(500,500),(0,255,255),5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,'markpai',(10,500),font,4,(255,0,0),5)
plt.imshow(blank_img)
```
