```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# giving a wrong path
img = cv2.imread('../DATA/00-puppy.jpg')
img.shape
```

```python
plt.imshow(img)
```

```python
img_rgb= cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
img_gray.shape
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap = 'gray')
```

# Resize Images 

```python
img_rgb.shape
```

```python
img = cv2.resize(img_rgb,(1300,270))
plt.imshow(img)
```

# By Ratio

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
plt.imshow(new_img)
```

```python
%matplotlib qt
plt.imshow(new_img)
```

# Flipping Images 

```python
%matplotlib inline
```

```python
new_img1 = cv2.flip(new_img,0)
plt.imshow(new_img1)
```

```python
new_img1 = cv2.flip(new_img,1)
plt.imshow(new_img1)
```

```python
new_img1 = cv2.flip(new_img,-1)
plt.imshow(new_img1)
```
