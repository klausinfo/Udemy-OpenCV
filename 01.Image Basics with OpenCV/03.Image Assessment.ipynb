```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../DATA/dog_backpack.png")
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
new_img = img_rgb
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
plt.imshow(img_rgb)
```

```python
cv2.rectangle(img_rgb,(200,380),(600,700),(0,255,0),5)
plt.imshow(img_rgb)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
vertices = np.array([[250,700],[375,400],[500,700]],np.int32)
pts = vertices.reshape((-1,1,2))
cv2.polylines(img_rgb,[pts],isClosed = True,color = (0,255,0),thickness = 20)
plt.imshow(img_rgb)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
vertices =np.array([[250,700],[425,400],[600,700]],np.int32)
pts = vertices.reshape((-1,1,2))
cv2.fillPoly(img_rgb,[pts],(255,0,0))
plt.imshow(img_rgb)
```

```python
import cv2
import numpy as np

def draw_circle(event,x,y,flags,param) :
    if event == cv2.EVENT_RBUTTONDOWN :
        cv2.circle(img,(x,y),100,(0,0,255),10)
        
img = cv2.imread("../DATA/dog_backpack.png")
cv2.namedWindow('circles')
cv2.setMouseCallback('circles',draw_circle)

while True :
    cv2.imshow('circles',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()        
```
